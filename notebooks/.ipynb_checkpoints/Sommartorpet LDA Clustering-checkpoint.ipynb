# Sommartorpet LDA clustering for topic modelling

```python
import pandas as pd
import re
import nltk
import gensim
import spacy
from nltk.stem.porter import *
from tqdm import tqdm
tqdm.pandas()
```

```python
nltk.download('punkt')
nltk.download('stopwords')
sw = nltk.corpus.stopwords.words('swedish')
#print(sw)
```

```python
path = './data/conv_data.csv'
df = pd.read_csv(path, usecols=['authorID', 'msg'])
df.info()
```

```python
""" Filter out NULL msgs """

df = df[df['msg'].notnull()]
df.info()
```

```python
""" To Lowercase """
df['msg'] = df['msg'].progress_apply(lambda x: x.lower())
```

```python
""" Filter out non-alphanumerics """
df['msg'] = df['msg'].progress_apply(lambda x: re.sub(r'[^a-öA-Ö-/\s:]', '', x))
```

```python
""" PREPARE DATA """
nlp = spacy.load("sv_core_news_sm", disable=["parser", "ner"])
df['nlp'] = df['msg'].progress_apply(lambda msg: [x.lemma_ for x in nlp(msg) if x.lemma_ not in sw])
```

```python
from wordcloud import WordCloud
lst = []
df['nlp'].progress_apply(lambda tokens: lst.append(' '.join(tokens))) # this is dumb

long_string = ','.join(lst)

# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=1000, contour_width=3, contour_color='steelblue')

# Generate a word cloud
wordcloud.generate(long_string)

# Visualize the word cloud
wordcloud.to_image()
```
